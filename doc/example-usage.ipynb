# Example usage

Run the FRAGMENT-MNP model is a two-step process. First, the model must be initialised by passing it config and input data. Example config and data is given in the `fragmentmnp.examples` module, which is used here. Then the `FragmentMNP.run()` method runs the model and returns a `FMNPOutput` named tuple with the output data.

```python
from fragmentmnp import FragmentMNP
from fragmentmnp.examples import minimal_config, data

# Create the model and pass it config and data. minimal_config is
# an example of a config dict with only required values given.
# full_config is an example of a config dict with all values given
fmnp = FragmentMNP(minimal_config, data)
# Run the model
output = fmnp.run()
```

The returned `FMNPOutput` object contains a timeseries `t`, particle number concentrations for each size class `n` and particle number concentration lost from the smallest size class `n_loss`. We can easily plot these to show model outputs:

```python
import matplotlib.pyplot as plt

# We transpose the particle number concentrations `n` to be able to pass
# all the size classes to matplotlib in one line, but we could just have
# loop through the size classes
plt.plot(output.t, output.n.T)
# fmnp.psd is the particle size distribution used in the model (specified
# in the config dict), so we can use that to put the size class diameters
# in the legend
plt.legend([f'{d} m' for d in fmnp.psd])
plt.show()
```
